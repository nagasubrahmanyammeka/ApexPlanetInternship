```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load Clean Dataset

df = pd.read_csv("cleaned_new_data.csv")

print("First 5 Rows:")
print(df.head())
```

```python
# 2. Descriptive Statistics
print("\n Summary Statistics:")
print(df.describe())

print("\n Stress Level Distribution:")
print(df['stress_level'].value_counts())

print("\n Platform Distribution:")
print(df['platform'].value_counts())
```

```python
# 3. Univariate Analysis


# Histogram - Screen Time
plt.figure()
df['daily_screen_time_hours'].hist()
plt.title("Screen Time Distribution")
plt.xlabel("Hours")
plt.ylabel("Frequency")
plt.show()

# Bar Chart - Platform Usage
plt.figure()
df['platform'].value_counts().plot(kind='bar')
plt.title("Platform Usage")
plt.xlabel("Platform")
plt.ylabel("Users")
plt.show()
```

```python
# 4. Bivariate Analysis


# Screen Time vs Productivity
plt.figure()
sns.scatterplot(x='daily_screen_time_hours', y='productivity_score', data=df)
plt.title("Screen Time vs Productivity")
plt.show()

# Sleep vs Stress Level
plt.figure()
sns.boxplot(x='stress_level', y='sleep_hours', data=df)
plt.title("Sleep vs Stress Level")
plt.show()

# Exercise vs Productivity
plt.figure()
sns.scatterplot(x='exercise_minutes', y='productivity_score', data=df)
plt.title("Exercise vs Productivity")
plt.show()
```

```python
# 5. Correlation Heatmap

plt.figure()
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# 6. Business Insights (Print)

print("\n Average Productivity by Stress Level:")
print(df.groupby('stress_level')['productivity_score'].mean())

print("\n Average Productivity by Activity Level:")
print(df.groupby('activity_level')['productivity_score'].mean())

print("\n Average Screen Time by Platform:")
print(df.groupby('platform')['daily_screen_time_hours'].mean())
```
